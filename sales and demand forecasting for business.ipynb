```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

df = pd.read_csv("Sample - Superstore.csv", encoding='latin1')
df.head()

# ð Sales Forecasting Project

This project predicts future sales using historical data to support business decisions.

```python
## ð§¹ Data Cleaning
```

## ð§¹ Data Cleaning


```python
df = df.dropna()
df['Order Date'] = pd.to_datetime(df['Order Date'])
df = df.sort_values('Order Date')
```

## ð Sales Trend Over Time

```python
sales = df.groupby('Order Date')['Sales'].sum().reset_index()

plt.figure(figsize=(10,5))
plt.plot(sales['Order Date'], sales['Sales'])
plt.title("Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()
```

## ð Monthly Sales Trend

```python
monthly_sales = df.groupby(df['Order Date'].dt.month)['Sales'].sum()

monthly_sales.plot(kind='bar')
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

## ð¤ Model Building

```python
from sklearn.linear_model import LinearRegression

sales['days'] = (sales['Order Date'] - sales['Order Date'].min()).dt.days

X = sales[['days']]
y = sales['Sales']

model = LinearRegression()
model.fit(X, y)
```

## ð® Sales Forecast

```python
future_days = pd.DataFrame({
    'days': range(sales['days'].max()+1, sales['days'].max()+31)
})

predictions = model.predict(future_days)
```

```python
plt.figure(figsize=(10,5))
plt.plot(sales['days'], sales['Sales'], label="Actual")
plt.plot(future_days['days'], predictions, label="Forecast")
plt.legend()
plt.title("Sales Forecast")
plt.show()
```

```python
future_days['Date'] = pd.date_range(start=sales['Order Date'].max(), periods=30, freq='D')
future_days['Predicted Sales'] = predictions

future_days[['Date', 'Predicted Sales']]
```

Monthly analysis reveals seasonal demand patterns.

The forecast predicts continued growth in the next 30 days.
